**Handling Missing Values**

```python
df.isnull().sum()[df.isnull().sum() != 0]
```

```python
# search related columns
search_cols = [
    'Search Conducted',
    'Search Disposition',
    'Search Outcome',
    'Search Reason',
    'Search Reason For Stop',
    'Search Type',
    'Search Arrest Reason'
]

df[search_cols] = df[search_cols].fillna('Not Applicable')
```

```python
df['Location'].fillna('unknown',inplace=True)
df['Description'].fillna('Not Recorded',inplace=True)
df['Color'].fillna('Unknown', inplace=True)
df['Article'].fillna('Not Recorded', inplace=True)
```

```python
df['State'].fillna(df['State'].mode()[0],inplace=True)
df['Make'].fillna(df['Make'].mode()[0],inplace=True)
df['Model'].fillna(df['Model'].mode()[0],inplace=True)
df['Driver City'].fillna(df['Driver City'].mode()[0],inplace=True)
df['DL State'].fillna(df['DL State'].mode()[0],inplace=True)
df['Driver State'].fillna(df['Driver State'].mode()[0],inplace=True)
#Less missing values so i replaced it with mode
```

```python
median_year = df['Year'].median()
df['Year'].fillna(median_year, inplace=True)
```

## ***Feature Engineering***

```python
df['SeqID'].describe()
```

```python
df["SeqID"] = df["SeqID"].astype(str)
# SeqID as an ID, not a number
```

```python
df['Date Of Stop'].unique()
```

```python
df["Date Of Stop"] = pd.to_datetime(df["Date Of Stop"], errors="coerce")
# Convert 'Date Of Stop' column to datetime format
# Invalid or unparseable dates will be set as NaT (missing value)
```

```python
df["Date Of Stop"].isna().sum()
```

```python
df['Stop Hour'] = pd.to_datetime(
    df['Time Of Stop'].astype(str)
).dt.hour
```

```python
df['Stop Hour']
```

```python
df['Time Of Stop'] = pd.to_datetime(
    df['Time Of Stop'].astype(str),
    format='%H:%M:%S',
    errors='coerce'
).dt.time
```

```python
df['Time Of Stop']
```

```python
df["Agency"] = df["Agency"].str.upper().str.strip()
```

```python
df['Agency'].describe()
```

```python
# Extract numeric district values from the SubAgency text column
# \d+ captures one or more digits from the string
# Convert result to nullable integer (Int64) to handle missing values
df['District Number'] = (
    df['SubAgency']
    .str.extract(r'(\d+)')
    .astype('Int64')
)
```

**.str.extract(r'(\d+)')**

Uses regular expression

\d+  one or more digits

Extracts the first number found in the string

```python
df['District Number']
```

```python
cols = ['SubAgency', 'Description', 'Location','Search Reason','Model']

df[cols] = df[cols].apply(
    lambda x: x.astype(str).str.strip().str.lower()
)
```

```python
df['Latitude']
```

```python
df.loc[
    (df['Longitude'] < -180) | (df['Longitude'] > 180),
    'Longitude'
]
```

```python
df.loc[
    (df['Longitude'] < -65) | (df['Longitude'] > -125),
    'Longitude'
]
```

```python
bool_cols = [
    "Accident",
    "Belts",
    "Personal Injury",
    "Property Damage",
    "Fatal",
    "Commercial License",
    "HAZMAT",
    "Commercial Vehicle",
    "Alcohol"
]

for col in bool_cols:
    print(f"\nValue counts for {col}")
    print(df[col].value_counts(dropna=False))

```

```python
df[bool_cols] = df[bool_cols].replace({'Yes': True, 'No': False})
```

```python
df[bool_cols].dtypes
```

```python
df['Work Zone']=df['Work Zone'].replace({'Yes':True,'No':False})
```

```python
df['Work Zone'].dtypes
```

```python
df['State']=df['State'].str.upper()
```

```python
df['VehicleType'].unique()
```

```python
# Clean and standardize the VehicleType text column
# Convert to string to safely apply text operations
# Normalize spacing around hyphens (e.g., "Car-Truck"  "Car - Truck")
# Remove leading and trailing spaces
df['VehicleType'] = (
    df['VehicleType']
    .astype(str)
    .str.replace(r'\s*-\s*', ' - ', regex=True)
    .str.strip()
)
```

####****str.replace(r'\s*-\s*', ' - ', regex=True)**

Regex breakdown:

\s*  zero or more spaces

-  hyphen

\s*  zero or more spaces

```python
# Extract numeric vehicle type code from the beginning of VehicleType
# ^\d+ captures digits only if they appear at the start of the string
# Convert to nullable integer (Int64) to allow missing values
df['VehicleType_Code'] = (
    df['VehicleType']
    .str.extract(r'^(\d+)')
    .astype('Int64')
)
```

### ***str.extract(r'^(\d+)')***


Regex breakdown:

^  start of the string

\d+  one or more digits

```python
df['VehicleType_Code'].unique()
```

```python
# Remove numeric vehicle code from VehicleType text
# ^\d+\s*-\s* matches leading numbers followed by a dash
# Strip extra spaces and convert to lowercase for consistency
df['VehicleType_Category'] = (
    df['VehicleType']
    .str.replace(r'^\d+\s*-\s*', '', regex=True)
    .str.strip()
    .str.lower()
)
```

####****str.replace(r'^\d+\s*-\s*', '', regex=True)**

^  start of string

\d+  one or more digits

\s*-\s*  dash with optional spaces

```python
df['VehicleType_Category'].unique()
```

```python
df.Year
```

```python
df.Year=df['Year'].astype(int)
```

```python
# Identify invalid years outside the realistic range
invalid_years = df.loc[
    (df['Year'] < 1950) | (df['Year'] > 2026),
    'Year'
].unique()

# Replace invalid years with missing values (NaN)
df.loc[
    (df['Year'] < 1950) | (df['Year'] > 2026),
    'Year'
] = pd.NA
```

```python
df.Color.value_counts()
```

```python
df['Color'] = df['Color'].astype(str).str.strip().str.upper()
```

```python
# Standardize vehicle color values by merging similar shades
# This reduces duplicate categories like "BLUE, DARK" and "BLUE, LIGHT"
df['Color'] = df['Color'].replace({
    'BLUE, DARK': 'BLUE',
    'BLUE, LIGHT': 'BLUE',
    'GREEN, DK': 'GREEN',
    'GREEN, LGT': 'GREEN'
})
```

```python
df['Violation Type'].value_counts()
```

```python
df['Charge'].value_counts()
```

```python
df['Charge'] = df['Charge'].astype(str).str.strip().str.upper()
```

```python
# Group all charges by traffic stop (SeqID)
# Remove duplicate charges per stop
# Sort charges alphabetically and join them into a single string
charges_per_stop = (
    df.groupby('SeqID')['Charge']
    .apply(lambda x: ', '.join(sorted(set(x))))
    .reset_index(name='Charges_List')
)
```

```python
charges_per_stop
```

```python
df['Article'].value_counts()
```

```python
# Convert Article column to string, remove extra spaces,
# and standardize text to Title Case
df['Article'] = df['Article'].astype(str).str.strip().str.title()
```

```python
df['Driver City'].astype('str').str.strip().str.lower()
```

```python
df['DL State'].unique()
```

```python
df['DL State'] = df['DL State'].astype(str).str.strip().str.upper()
```

```python
df['Arrest Type'].value_counts()
```

```python
df['Arrest Type'] = (df['Arrest Type'].astype(str).str.strip())
```

```python
# Extract the arrest type code (single capital letter) from the start of Arrest Type
# ^([A-Z]) captures the first uppercase letter at the beginning of the string
df['Arrest_Type_Code'] = (
    df['Arrest Type']
    .str.extract(r'^([A-Z])')
)
```

***str.extract(r'^([A-Z])')***

^  start of string

[A-Z]  one uppercase letter

()  capture group (this is what gets returned)

```python
# Remove arrest type code (letter + dash) from Arrest Type text
# ^[A-Z]\s*-\s* matches a leading capital letter followed by a dash
# Convert description to lowercase for consistency
df['Arrest_Type_Desc'] = (
    df['Arrest Type']
    .str.replace(r'^[A-Z]\s*-\s*', '', regex=True)
    .str.lower()
)
```

****str.replace(r'^[A-Z]\s*-\s*', '', regex=True)***

^  start of string

[A-Z]  one capital letter

\s*-\s*  dash with optional spaces

```python
df.Geolocation
```

```python
# Extract latitude and longitude from Geolocation text
# Handles formats like "(lat, lon)", "lat,lon", or "lat , lon"
geo = df['Geolocation'].astype(str).str.extract(
    r'\(?\s*([-+]?\d*\.?\d+)\s*,\s*([-+]?\d*\.?\d+)\s*\)?'
)

# Convert extracted latitude to numeric (invalid values become NaN)
df['Geo_Lat'] = pd.to_numeric(geo[0], errors='coerce')

# Convert extracted longitude to numeric (invalid values become NaN)
df['Geo_Lon'] = pd.to_numeric(geo[1], errors='coerce')
```

```python
df.loc[
    (df['Geo_Lat'] == 0) & (df['Geo_Lon'] == 0),
    ['Geo_Lat', 'Geo_Lon']
] = pd.NA
```

```python
df[['Latitude', 'Geo_Lat']].corr()
df[['Longitude', 'Geo_Lon']].corr()
```

```python
df.drop(columns=['Geolocation', 'Geo_Lat', 'Geo_Lon'], inplace=True)
```

```python
df.to_parquet('traffic_stops_clean.parquet', index=False)
```
